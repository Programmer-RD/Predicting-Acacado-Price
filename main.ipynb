```python
def object_to_int(info,coloum):
    index = -1
    info_dict = {}
    for coloum_info in info[coloum]:
        if coloum_info not in info_dict:
            index = index + 1
            info_dict[coloum_info] = index
    all_dict = []
    for coloum_info in info[coloum]:
        all_dict.append(info_dict[coloum_info])
    return all_dict,info_dict
```

```python
import json
with open('type_info_json.json','w') as json_file:
#     json.load(json_file)
    json.dump(object_to_int(data,'type')[1],json_file)
data['type'] = object_to_int(data,'type')[0]
```
